```python
import csv
import re
pos_file = "/media/data/hotels/booking_v2/processed/extracted_ap/aspect_positive.cont.csv"
neg_file = "/media/data/hotels/booking_v2/processed/extracted_ap/aspect_negative.cont.csv"
```

```python
def read_csv(fname):
    data = []
    with open(fname, "r") as f:
        csvreader = csv.reader(f)
        for line in csvreader:
            data.append(line)
    return data
```

```python
pos_data = read_csv(pos_file)
sbut_clauses = [line[0] for line in pos_data if re.match(r"but", line[0])]
ibut_clauses = [line[0] for line in pos_data if not re.match(r"but", line[0]) and line[1]=="but"]
```
